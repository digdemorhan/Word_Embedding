```python
#Verisetini yazdırma
import pandas as pd
df = pd.read_csv('TRSAv1.csv')
pd.set_option('display.max_colwidth', None)
print(df.head(10))
print()
print(df.tail(10))
```

```python
#Verisetindeki her sınıftan rastgele 1000 örnek alma
positive_exmp = df[df['score'] == 'Positive'].sample(n=1000, random_state=42)
negative_exmp = df[df['score'] == 'Negative'].sample(n=1000, random_state=42)
neutral_exmp = df[df['score'] == 'Neutral'].sample(n=1000, random_state=42)

#Örnekleri birleştirme
exmp = pd.concat([positive_exmp, negative_exmp, neutral_exmp]).reset_index(drop=True)

#Verisetinin yeni halinin ilk 10 verisi
exmp.head(10)
```

## **Preprocessing**

```python
#Lower Casing
exmp['review'] = exmp['review'].str.lower()
exmp[['id','score','review']].head(10)
```

```python
exmp[['id','score','review']].tail(10)
```

```python
exmp[['id','score','review']].iloc[1000:1010]
```

```python
#Removal of Punctionations
import string

def remove_punctination(text):
  return text.translate(str.maketrans('','',string.punctuation))

exmp['review'] = exmp['review'].apply(remove_punctination)
exmp[['id','score','review']].head(10)
```

```python
#Removal of Stopwords
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')

stop_words = set(stopwords.words('turkish'))
exmp['review'] = exmp['review'].apply(lambda x: ' '.join([word for word in x.split() if word not in stop_words]))

exmp[['id','score','review']].iloc[1000:1010]
```

```python
#Lemmatization
!pip install zeyrek
from zeyrek import MorphAnalyzer

morph_analyzer = MorphAnalyzer()

def lemmatize_text(text):
    words = text.split()
    lemmatized_words = [morph_analyzer.lemmatize(word)[0][0] for word in words]
    return ' '.join(lemmatized_words)

exmp['review'] = exmp['review'].apply(lemmatize_text)
exmp[['id','score','review']].head(10)
```

```python
#Removal of Emojis
import re

def remove_emojis(text):
    emoji_pattern = re.compile("[\U0001F600-\U0001F64F"  # Emoticons
                                 "\U0001F680-\U0001F6FF"
                                 "\U0001F800-\U0001F8FF"
                                 "\U00002702-\U000027B0"
                                 "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)

exmp['review'] = exmp['review'].apply(remove_emojis)

exmp[['id', 'score', 'review']].head(10)
```

## **Word Embedding**

```python
from gensim.models import Word2Vec
import nltk

cumleler = [nltk.word_tokenize(cumle) for cumle in exmp['review']]

model = Word2Vec(cumleler, vector_size=50, window=3, min_count=1, workers=4)

model.save("/content/drive/MyDrive/WordEmbedding.model")

from gensim.models import Word2Vec

model = Word2Vec.load("/content/drive/MyDrive/WordEmbedding.model")

cilt_vektoru = model.wv['cilt']
print("Cilt Kelimesinin Vektörü:")
print(cilt_vektoru)
print()
benzer_kelimeler = model.wv.most_similar('cilt', topn=5)
print("Cilt Kelimesi İçin Benzer Kelimeler:")
for kelime in benzer_kelimeler:
    print(kelime)
print()
bakim_vektoru = model.wv['bakım']
print("Bakım Kelimesinin Vektörü:")
print(bakim_vektoru)
print()
benzer_kelimeler = model.wv.most_similar('bakım', topn=5)
print("Bakım Kelimesi İçin Benzer Kelimeler:")
for kelime in benzer_kelimeler:
    print(kelime)
print()

try:
    analoji_sonuc = model.wv.most_similar(positive=['bakım', 'cilt'], negative=['ürün'], topn=5)
    print("Analoji Sonucu:")
    for kelime in analoji_sonuc:
        print(kelime)
except KeyError as e:
    print(f"Kelime bulunamadı: {e}")
print()
try:
    analoji_sonuc = model.wv.most_similar(positive=['şampuan', 'temizlik'], negative=['kuru'], topn=5)
    print("Analoji Sonucu:")
    for kelime in analoji_sonuc:
        print(kelime)
except KeyError as e:
    print(f"Kelime bulunamadı: {e}")
print()
try:
    analoji_sonuc = model.wv.most_similar(positive=['etkili', 'krem'], negative=['kötü'], topn=5)
    print(f"'etkili + krem - kötü' işleminin sonucu:")
    for kelime in analoji_sonuc:
        print(f"{kelime[0]} (Benzerlik Skoru: {kelime[1]})")
except KeyError as e:
    print(f"Kelime bulunamadı: {e}")
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import nltk

exmp = {
    'review': [
        "Bu yüz kremi çok iyi bir ürün.",
        "Kremin etkisi harika, cildim pırıl pırıl oldu.",
        "Kötü bir etki yaratmadı ama çok da etkili değil.",
        "Hemen sonuç aldım, mükemmel bir ürün.",
        "Cildimde kötü bir alerjiye neden oldu."
    ]
}

count_vectorizer = CountVectorizer()
count_vectorized = count_vectorizer.fit_transform(exmp['review'])

kelimeler_count = count_vectorizer.get_feature_names_out()
sayilar_count = count_vectorized.toarray().flatten()

print("CountVectorizer Sonuçları:")
for kelime, sayi in zip(kelimeler_count, sayilar_count):
    print(f"{kelime}: {sayi}")
print()
tfidf_vectorizer = TfidfVectorizer()
tfidf_vectorized = tfidf_vectorizer.fit_transform(exmp['review'])

kelimeler_tfidf = tfidf_vectorizer.get_feature_names_out()
skorlar_tfidf = tfidf_vectorized.toarray().flatten()

print("TfidfVectorizer Sonuçları:")
for kelime, skor in zip(kelimeler_tfidf, skorlar_tfidf):
    print(f"{kelime}: {skor}")
```
