```python
#Verisetini yazdÄ±rma
import pandas as pd
df = pd.read_csv('TRSAv1.csv')
pd.set_option('display.max_colwidth', None)
print(df.head(10))
print()
print(df.tail(10))
```

```python
#Verisetindeki her sÄ±nÄ±ftan rastgele 1000 Ã¶rnek alma
positive_exmp = df[df['score'] == 'Positive'].sample(n=1000, random_state=42)
negative_exmp = df[df['score'] == 'Negative'].sample(n=1000, random_state=42)
neutral_exmp = df[df['score'] == 'Neutral'].sample(n=1000, random_state=42)

#Ã–rnekleri birleÅŸtirme
exmp = pd.concat([positive_exmp, negative_exmp, neutral_exmp]).reset_index(drop=True)

#Verisetinin yeni halinin ilk 10 verisi
exmp.head(10)
```

## **Preprocessing**

```python
#Lower Casing
exmp['review'] = exmp['review'].str.lower()
exmp[['id','score','review']].head(10)
```

```python
exmp[['id','score','review']].tail(10)
```

```python
exmp[['id','score','review']].iloc[1000:1010]
```

```python
#Removal of Punctionations
import string

def remove_punctination(text):
  return text.translate(str.maketrans('','',string.punctuation))

exmp['review'] = exmp['review'].apply(remove_punctination)
exmp[['id','score','review']].head(10)
```

```python
#Removal of Stopwords
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')

stop_words = set(stopwords.words('turkish'))
exmp['review'] = exmp['review'].apply(lambda x: ' '.join([word for word in x.split() if word not in stop_words]))

exmp[['id','score','review']].iloc[1000:1010]
```

```python
#Lemmatization
!pip install zeyrek
from zeyrek import MorphAnalyzer

morph_analyzer = MorphAnalyzer()

def lemmatize_text(text):
    words = text.split()
    lemmatized_words = [morph_analyzer.lemmatize(word)[0][0] for word in words]
    return ' '.join(lemmatized_words)

exmp['review'] = exmp['review'].apply(lemmatize_text)
exmp[['id','score','review']].head(10)
```

```python
#Removal of Emojis
import re

def remove_emojis(text):
    emoji_pattern = re.compile("[\U0001F600-\U0001F64F"  # Emoticons
                                 "\U0001F680-\U0001F6FF"
                                 "\U0001F800-\U0001F8FF"
                                 "\U00002702-\U000027B0"
                                 "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)

exmp['review'] = exmp['review'].apply(remove_emojis)

exmp[['id', 'score', 'review']].head(10)
```
